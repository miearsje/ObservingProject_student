4. **Problem** -  There are lines in common between M57 and Vega but they are in absorption in Vega and in Emission in M57. 
**(a)** Identify the lines in common.  *Note that there may be small wavelength shits between the two spectrum because of imperfect wavelength calibration.*
    
**(b)** Explain why they are in emission in one spectrum and in absorption in the other.  Use your knowledge of Kirkhoffs laws. *Note - The line at 6880$\mathring A$ is from the Earth's atmosphere and should be ignored*

**answer here**

**(a)** Lines in common:
<ul>
    <li>H$_{\epsilon}$ at 3970 $\mathring A$</li> 
    <li>H$_{\delta}$ at 4103 $\mathring A$</li> 
    <li>H$_{\gamma}$ at 4341 $\mathring A$</li> 
    <li>H$_{\beta}$ at 4861 $\mathring A$</li> 
    <li>H$_{\alpha}$ at 6563 $\mathring A$</li>
</ul>

**(b)** When an electron moves to a lower energy level, a photon is emitted. This is what we are seeing when looking at spectra and find more photons than expected. Kirchoff's second law states that hot gas against a dark background will produce an emission line spectrum. <br>
Conversely, absorption happens when the electrons absorb the photons, moving to higher energy levels, causing less photons observed than expected. This brings us to Kirchoff's third law, which is when light is absorbed by a cool gas, causing absorption line spectra. <br>
The absorption happening in Vega's spectrum would therefore be due to cool hydrogen gas between us and Vega absorbing some light, whereas the emission happening in M57 would be caused by hot hydrogen gas between us and M57 emitting light.
